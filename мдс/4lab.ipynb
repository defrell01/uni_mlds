4. Создание нового датафрейма с физическими показателями


```python
df_phys = data[['temperature_deg_c', 'ph', 'salinity_psu']]
df_phys = df_phys.drop_duplicates().dropna()
df_phys
```

```python
df_empo = data['empo_3']
df_empo
```

4. Объединение через merge по индексам


```python
df_merged = df_seqs.merge(df_phys, how='inner',
                          right_index=True, left_index=True)
df_merged
```

6. Объединение через join()


```python
df_merged = df_merged.join(df_empo)
df_merged
```

```python
df_merged.info()
df_merged.describe()
```
